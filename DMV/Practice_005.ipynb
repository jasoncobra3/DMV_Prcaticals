```python
def top_and_bottom_10(indicator="so2"):
    fig, ax =plt.subplots(2,1,figsize=(20,12))
    ind=df[[indicator , "state"]].groupby("state",as_index=False).median().sort_values(by=indicator , ascending=False)
    top_10=sns.barplot(x="state" ,y=indicator, data=ind[:10] , ax=ax[0] ,color="red" )
    top_10.set_title("Top 10 States by {} (1990-2016)".format(indicator))
    top_10.set_ylabel("{} (g/m3)".format(indicator))
    top_10.set_xlabel("State")
    bottom_10=sns.barplot(x="state" ,y=indicator, data=ind[-10:] , ax=ax[1] ,color="green" )
    bottom_10.set_title("Bottom 10 States by {} (1991-2016)".format(indicator))
    bottom_10.set_ylabel("{} (g/m3)".format(indicator))
    bottom_10.set_xlabel("State")
```

```python
top_and_bottom_10("so2")
top_and_bottom_10("no2")
```

```python
def highest_level_record(indicator=""):
    plt.figure(figsize=(20,10))
    ind = df[[indicator, 'state']].groupby('state', as_index=False).max() 
    highest = sns.barplot(x='state', y=indicator, data=ind)
    highest.set_title("Highest ever {} levels recorded by state".format(indicator))
    plt.xticks(rotation=90) 
highest_level_record("so2")
highest_level_record("no2")
```

```python
def type_avg(indicator=""):
    # Filter to select only numeric columns in value_col5 and add "state"
    numeric_cols = df[value_col5].select_dtypes(include='number').columns.tolist()
    type_avg = df[numeric_cols + ["state"]].groupby("state").mean()
    
    # Plot based on indicator
    if indicator and indicator in type_avg.columns:
        # Plot specific indicator if it exists in the dataframe
        type_avg[indicator].plot(kind="bar", color='skyblue')
        plt.title("Pollutant Average by State for {}".format(indicator))
    else:
        # Plot all indicators if none specified or if the indicator is invalid
        type_avg.plot(kind="bar", figsize=(10, 6), colormap='viridis')
        plt.title("Pollutant Average by State for All Indicators")

    plt.xticks(rotation=90)
    plt.xlabel("State")
    plt.ylabel("Average Value")
    plt.show()
```

```python
type_avg("no2")
```

```python
def location_avgs(state, indicator="so2"):
    # Select only numeric columns from value_col5 for aggregation
    numeric_cols = df[value_col5].select_dtypes(include='number').columns.tolist()
    
    # Group by 'state' and 'location' with only numeric columns
    locs = df[numeric_cols + ['state', 'location']].groupby(['state', 'location']).mean()
    
    # Filter data for the specified state
    state_avgs = locs.loc[state].reset_index()

    # Plotting
    sns.barplot(x='location', y=indicator, data=state_avgs) 
    plt.title("Location-wise average for {} in {}".format(indicator, state)) 
    plt.xticks(rotation=90)
    plt.xlabel("Location")
    plt.ylabel("Average {}".format(indicator))
    plt.show()


location_avgs("Bihar", "no2")
```
