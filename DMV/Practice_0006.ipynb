```python
df=combined_pivot.copy()
plt.figure(figsize=(12, 8))
bottom = None  # Initialize bottom for stacking

# Plot each category on top of the previous
for category in df.columns:
    if bottom is None:
        plt.bar(df.index, df[category], label=category)  # Base layer
        bottom = df[category]
    else:
        plt.bar(df.index, df[category], bottom=bottom, label=category)  # Stacked layer
        bottom += df[category]  # Update bottom for the next category

# Labeling
plt.xlabel('Shopping Mall')
plt.ylabel('Total Sales')
plt.title('Sales by Category in Each Shopping Mall')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')

# Display plot
plt.tight_layout()
plt.show()
```

```python
df=combined_pivot.copy()
bottom=None
plt.figure(figsize=(12,8))

for category in df.columns:
    if bottom is None:
        plt.bar(df.index , df[category] , label=category)
        bottom=df[category]
    else:
        plt.bar(df.index , df[category] , bottom=bottom , label=category)
        bottom += df[category]


plt.title("Sales by Category in Each mall")
plt.xlabel("Shopping Mall")
plt.ylabel("Total sales")
plt.xticks(rotation=45)
plt.legend(title="category" , loc="upper right")
plt.tight_layout()
plt.show()
```
