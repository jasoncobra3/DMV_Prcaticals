```python
data=df.drop(["N","P","K","ph","rainfall","label"], axis=1)
```

```python
data.head()
```

```python
x=data["temperature"]
y=data["humidity"]
```

```python
x.skew()
```

```python
y.skew()
```

```python
sns.histplot(x , color="purple")
plt.title("Histogram for Temperature")
plt.show()
```

```python
sns.boxplot(x)
plt.title("Boxplot for Temperature")
plt.show()
```

```python
sns.histplot(y , color="yellow")
plt.title("Histogram for Humidity")
plt.show()
```

```python
sns.boxplot(y)
plt.title("Boxplot for Humidity")
plt.show()
```

```python
plt.plot(x ,color="red")
plt.plot(y , color="blue")
plt.title("Line plot for Temperature and humidity")
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Scatter Plot for Temp v/s Humidity")
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.show()
```
